Now we see that the Cabin columns has information on the deck, But it has serval nan values , so we'll have to drope them

```python
# First we'll drop the NAN values and create a new object, deck  
deck = data['Cabin'].dropna()
deck
```

```python
# Let's grab that letter for deck level with a simple for loop 
levels = []
for level in deck:
    levels.append(level[0])

cabin_df = pd.DataFrame(levels)
cabin_df.columns = ['Cabin']
cabin_df.sort_values(by='Cabin', inplace = True)
sns.catplot(x='Cabin', data=cabin_df, kind='count', palette='winter_d')
```

```python
# "T" deck value there which does't make sence, we can deop it out with the following code:
cabin_df = cabin_df[ cabin_df.Cabin != 'T']
sns.catplot(x='Cabin', data=cabin_df, kind='count', palette ='summer')
```

```python
# Lets start by adding a new column to define alone 
# We'll add the parent / child column with the sibsp column

data['Alone'] = data.SibSp + data.Parch
data.Alone
```

Now we Know that if the Alone Column is anything But 0, the passanger had family aboard and wasn't alone. Som let's
Change the columns now so that value is greater then 0, we know the passenger was with his/her family, otherwise
they were alone.

```python
# Look For > 0 or == 0 to set alone status
#data.loc[data['Alone'] > 0, 'Alone'] ='with Family'
#data.loc[data['Alone'] == 0,'Alone'] = 'Alone'
#Lets check to make sure it worked

data.loc[data['Alone'] > 0, 'Alone'] = 'with Family'
data.loc[data['Alone'] == 0, 'Alone'] = 'Alone'
```

#5. What factore helping someone survive the sinking?

```python
# let's start by creating a new column for legibility purposes through mapping 
data['Survivor'] = data.Survived.map({0:'No', 1:'Yes'})

#Let's just get a quick overall view of Survived vs died.
sns.catplot(x='Survivor', data=data, kind='count')
```

So quite a few more people died than those who survived. Let's see if the class of the passengers had an effect on their survival rate, since the movie Titanic popularized the notion that the 3rd class passengers did not do as well as their 1st and 2nd class counterparts.

```python
# Lets use a factor plot again, But now considering class 
sns.catplot(x='Pclass', y='Survived', data=data, kind='point')
```

Look like survival rates for the 3rd class are substantially lower! But maybe this effect is being caused by the large amount of men in the 3rd class in combination with the women and children first policy. Let's use 'hue' to get a clearer picture on this.

```python
# Lets use a factor plot again, But Now considering class and gender
sns.catplot(x ='Pclass', y='Survived', data=data, hue='Person', kind='point')
```

 From this data it look like beign a male or beign in 3rd class were both not favourable for survived 
Enven regardless of class the result of beign a male in any class dramatically decreases
your chance of survival
But waht about age? Did being younger or older have an effect on survival rate?

```python
# Lets use a line plot an age versus survival 
#sns.Implot(x='Age', y='Survived', data=data)
# Let's use a linear plot on age versus survival
sns.lmplot(x='Age', y='Survived', data=data)
```

Looks like there is a general trend that the older the passenger was, the less likely they survived. Let's go ahead and use hue to take a look at the effect of class and age.

```python
# Lets use a linear plot on age versus using hue for class seperation
sns.lmplot(x='Age', y = 'Survived', hue='Pclass', data=data)
```

```python
# Let's use a linear plot on age varsus survival using hue for class Seperation 
generations = [10, 20, 40, 60, 80]
sns.lmplot(x='Age', y ='Survived', hue = 'Pclass', data=data , palette='winter',x_bins=generations)
```

Intreasting find on the older 1st class passengers! what about if we relate gender and age with the survival set?

```python
sns.lmplot(x='Age',y='Survived', hue = 'Sex', data=data, palette='winter', x_bins=generations)
```
